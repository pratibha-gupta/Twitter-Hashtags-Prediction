```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(ngram_range =(1,2),binary = True)
```

```python
X = vectorizer.fit_transform(df1['text'] )
#train = X['text'][ :900]
#test = X['text'][900:]
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,df1['hashtags'],test_size=0.2)
```

```python
#y_train = df1['hashtags'][ :900]
```

```python
Ytrain
```

```python
label_classes= dict()
for  labels in Ytrain:
    #for label in labels:
    if labels in label_classes:
        label_classes[labels] += 1
    else:    
        label_classes[labels] = 1
```

```python
label_classes
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mb = MultiLabelBinarizer(classes = sorted(label_classes.keys()))
Y = mb.fit_transform(df1['hashtags'])
```

```python
#Y_train = Y[:900]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.multiclass import OneVsRestClassifier
lr = LogisticRegression(solver='lbfgs')
clf = OneVsRestClassifier(lr)
clf.fit(Xtrain,Ytrain)
```

```python
#x_test = vectorizer.fit_transform(test)
```

```python
pred = clf.predict(Xtest)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
print("Test Accuracy:" + str(accuracy_score(Ytest,pred)))
```
